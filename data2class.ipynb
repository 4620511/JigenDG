```python
from torchvision.datasets import ImageFolder
```

```python
datasets = ["CALTECH", "LABELME", "PASCAL", "SUN"]
paths =  {dataset: "/home/enoon/data/images/VLCS/%s/test" % dataset for dataset in datasets}
list(paths.values())[0]
```

```python
class PathLoader(ImageFolder):
    def __getitem__(self, index):
        """
        Args:
            index (int): Index
        Returns:
            tuple: (sample, target) where target is class_index of the target class.
        """
        path, target = self.samples[index]
        return path, target
```

```python
def get_path_to_class(path_loader):
    lines = []
    for (path, label) in path_loader:
        lines.append("%s %d\n" % (path, label))
    return lines
```

```python
for k, v in paths.items():
    lines = get_path_to_class(PathLoader(v))
    with open("data/txt_lists/%s_test.txt" % k, "wt") as f:
        f.writelines(lines)
```
