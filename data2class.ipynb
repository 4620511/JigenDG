```python
from torchvision.datasets import ImageFolder
```

```python
pacs_datasets = ["art_painting", "cartoon", "photo", "sketch"]
pacs_paths =  {dataset: "/home/enoon/data/images/PACS/kfold/%s" % dataset for dataset in pacs_datasets}
sample_path = list(pacs_paths.values())[0]
```

```python
class PathLoader(ImageFolder):
    def __getitem__(self, index):
        """
        Args:
            index (int): Index
        Returns:
            tuple: (sample, target) where target is class_index of the target class.
        """
        path, target = self.samples[index]
        return path, target
```

```python
def get_path_to_class(path_loader):
    lines = []
    for (path, label) in path_loader:
        lines.append("%s %d\n" % (path, label))
    return lines
```

```python
for k, v in pacs_paths.items():
    lines = get_path_to_class(PathLoader(v))
    with open("data/txt_lists/%s_train.txt" % k, "wt") as f:
        f.writelines(lines)
```
